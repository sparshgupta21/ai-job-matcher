### Step 3: Configure API Tokens

**Required:** Get a free HuggingFace token at https://huggingface.co/settings/tokens

**Optional:** Get a free JSearch API key at https://rapidapi.com/letscrape-6bRBa3QguO5/api/jsearch

```
import os
from getpass import getpass

# Required: HuggingFace Token
hf_token = getpass('Enter your HuggingFace Token (hf_...): ')
os.environ['HF_TOKEN'] = hf_token
print('\u2705 HF Token set!')

# Optional: JSearch API Key
jsearch_key = getpass('Enter JSearch API Key (press Enter to skip): ')
if jsearch_key:
    os.environ['JSEARCH_API_KEY'] = jsearch_key
    print('\u2705 JSearch API Key set!')
else:
    print('\u26a0\ufe0f No JSearch key - using demo data mode')
```

### Step 4: Launch the App

After running this cell, click the **public URL** (the one with `gradio.live`) to open the app.

```
# Launch the AI Job Matcher!
from app import create_app

app = create_app()
app.launch(share=True, debug=False)
```

## Troubleshooting

| Issue | Solution |
|-------|----------|
| **Token error** | Make sure your HF token starts with `hf_` |
| **No jobs found** | Set a JSearch API key or use demo mode |
| **Model error** | Try a different model from the dropdown |
| **Timeout** | Free HF API has rate limits; wait and retry |
| **Import error** | Re-run the Install Dependencies cell |
